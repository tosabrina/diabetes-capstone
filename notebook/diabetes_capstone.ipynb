```python
# Find outliers using a function:
def find_outliers_IQR(no_quest_mark_df):

   q1=no_quest_mark_df.quantile(0.25)

   q3=no_quest_mark_df.quantile(0.75)

   IQR=q3-q1
    
   outliers = no_quest_mark_df[((no_quest_mark_df<(q1-1.5*IQR)) | (no_quest_mark_df>(q3+1.5*IQR)))]

   return outliers
```

```python
outliers = find_outliers_IQR(no_quest_mark_df[['time_in_hospital', 'num_lab_procedures', 'num_procedures', 'num_medications', 'number_outpatient', 'number_emergency', 'number_inpatient', 'number_diagnoses']])
```

# 4. Are there disparities in treatment outcomes based on demographic factors?

**NOTE: 
- use chi-squared test to compare observed outcomes with expected outcomes
- convert readmitted column into boolean variables

```python
# Crosstab for race vs. readmission:
race_readmit = pd.crosstab(diab_merge['race'], diab_merge['readmitted'])
chi2, p, dof, expected = stats.chi2_contingency(race_readmit)
print(f"Chi-square test p-value: {p}")

# Visualize readmission rates by race:
sns.countplot(x='race', hue='readmitted', data=diab_merge)
plt.title('Readmission Rates by Race')
plt.xlabel('Race')
plt.ylabel('Number of Readmissions')
plt.xticks(rotation=45)
plt.show()
```

```python
print(f"Interpretation of Results: We cannot conclude that there is a significant relationship between the variables, meaning the observed outcomes are more likely due to random variation and possibly skewed from lack of demographic diversification")
```
