```python
# Find outliers using a function:
def find_outliers_IQR(no_quest_mark_df):

   q1=no_quest_mark_df.quantile(0.25)

   q3=no_quest_mark_df.quantile(0.75)

   IQR=q3-q1
    
   outliers = no_quest_mark_df[((no_quest_mark_df<(q1-1.5*IQR)) | (no_quest_mark_df>(q3+1.5*IQR)))]

   return outliers
```

```python
outliers = find_outliers_IQR(no_quest_mark_df[['time_in_hospital', 'num_lab_procedures', 'num_procedures', 'num_medications', 'number_outpatient', 'number_emergency', 'number_inpatient', 'number_diagnoses']])
```
